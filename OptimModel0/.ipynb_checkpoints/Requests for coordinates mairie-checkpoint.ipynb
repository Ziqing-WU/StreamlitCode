```python
import pandas as pd
import requests
import numpy as np

file_path_pop_den = r"C:\Users\zwu\OneDrive - IMT Mines Albi\Documents\Data\DemandForecasting\Vehicles\Population\grille_densite_4niveau_detaille_2022.csv"

pop_den = pd.read_csv(file_path_pop_den)

def retrieve_coordinates(Code_INSEE):
    api_url = f'https://geo.api.gouv.fr/communes/{Code_INSEE}?fields=code&format=geojson&geometry=mairie'
    response = requests.get(api_url)
    if response.status_code == 200:
        return response.json()["geometry"]["coordinates"]
```

```python
pop_den["coordinates"] = np.zeros(pop_den.shape[0])
```

```python
for i in range(pop_den.shape[0]):
    pop_den.at[i, 'coordinates'] = retrieve_coordinates(pop_den.iloc[i]['Code Commune'])
    if i%100==0:
        print(i)
```
